```python
student = [{"name":"小美"},
           {"name":"阿土"}]
first_name = "二哥"
for student_dic in student:
    print(student_dic)
    if student_dic["name"] == first_name:
        print("找到了%s"%first_name)
        break
else:
    print("查无此人")
print("循环结束")
```
