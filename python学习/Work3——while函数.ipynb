# 赋值运算符


```python
i = 0
a = 0
while i <= 100:
    a = i + a
    i = i + 1
print(a)
```

```python
i = 0 
a = 0
while i <= 100:
    a = i + a
    i = i + 2
print(a)
```

```python
i = 0
a = 0
while i <= 100:
    if i % 2 == 0:
        a = a + i
        i = i + 1
    else:
        i = i + 1
print(a)
```

# 循环嵌套

```python
i = 1 

while i <= 5:
    print("*" * i)
    i = i + 1
```

```python
i = 1
while i < 6:
    j = 0
    while j < i:
        print("*", end = "")
        j = j + 1
    print("")
    i = i + 1
```

```python
i = 1
while i <= 9:
    j = 1
    while j <= i:
        result = i * j
        print("%d * %d = %d"%(j,i,result),end = " ")
        j = j + 1
    print("")
    i = i + 1
```

```python
i = 1
while i <= 9:
    j = 1
    while j <= i:
        result = i * j
        print("%d * %d = %d"%(j,i,result),end = "\t")
        j = j + 1
    print("")
    i = i + 1
```
