```python
xiaoming = {"name":"xiaoming","age":18,"height":178,"weight":140}#键是唯一的
xiaoming
```

# 字典的操作

### 1.取值

```python
xiaoming["name"]
```

```python
xiaoming["name123"] #在取值时键不在会报错
```

### 增改

```python
xiaoming["name"] = "小明"          #如果键存在，会改值
xiaoming
```

```python
xiaoming["gender"] = True        #如果键不存在，会增
xiaoming
```

### 删除

```python
xiaoming.pop("age")
print(xiaoming)
```

```python
xiaoming.pop("name")   #如果键值不存在
```

### 统计键值对数量

```python
print(len(xiaoming))
```

```python
xiaojun = {"name":"xiaojun"}
xiaoming.update(xiaojun)
xiaoming
```

### 合并字典

```python
xiaohua = {"name":"xiaohua"}     #如果原来的字典中含有新增加的键，那么新增加的键值会覆盖原来的键
xiaoming.update(xiaohua)
xiaoming
```

### 清空字典

```python
xiaoming.clear()
xiaoming
```

### 循环遍历

```python
xiaoming = {"name":"xiaoming","age":"18","height":"178","weight":"140"}
for k in xiaoming: #k是每次循环迭代的键名
    print("%s - %s"%(k,xiaoming[k]))
```

# 字典和列表的使用场景

```python
card_list = [{"name":"张三","qq":"12345","phone":"10000"},
             {"name":"李四","qq":"54321","phone":"10086"}]
for card_info in card_list:
    print(card_info)
```
