# 练习

```python
age = int(input("请输入你的年龄："))
if age >= 0 and age <= 120:
    print("输入正确")
else:
    print("输入错误")
```

```python
python_score = 20
c_score = 100
if python_score > 60 or c_score > 60:
    print("成绩合格")
else:
    print("重考")
```

```python
holiday_name = input("今天是什么节")
if holiday_name == "情人节":
    print("买玫瑰/看电影")
elif holiday_name == "平安夜":
    print("买苹果/吃大餐")
elif holiday_name == "生日":
    print("买蛋糕")
else:
    print("其他的日子每天都是节日啊")
    
```

# 综合训练（修改为电脑随机出拳）

```python
import random

player = int(input("你想出什么：石头（1），剪刀（2），布（3）"))
computer  = random.randint(1,3)
print("电脑出的是%d"%computer)

if ((player == 1 and computer == 2)
        or (player == 2 and computer == 3)
        or (player == 3 and  computer == 1)):
    print("电脑真是弱爆了")
elif player == computer:
    print("你和电脑不相上下")
else:
    print("我不服，改日再战")
```
