```python
x = -121
string = str(x)
a = list(string)
b = []
for i in range(len(a)-1,-1,-1):
    b.append(a[i])

if a == b:
    print("true")
else:
    print("false")
```

```python
def solution(nums) :
    
    a = []
    for el in nums:
        if a.count(el) < 1:
            a.append(el)
    return a
c = [0,1,3,3,3,4]
solution(c)
```
