```python
# 函数的定义
def mutiple_table():
    i = 1
    while i <= 9:
        j = 1
        while j <= i:
            result = i * j
            print("%d * %d = %d"%(j,i,result),end = "\t")
            j = j + 1
        print("")
        i = i + 1
```

```python
def print_line(char,length):
    print(char * length)
def print_lines(char,length):
    i = 0
    while i < 5:
        print_line(char,length)
        i = i + 1
a = input("你想使用的符号时：")
b = int(input("你想使用的符号的长度："))
print_lines(a,b)
```
