```python
# 
def mutiple_table():
    i = 1
    while i <= 9:
        j = 1
        while j <= i:
            result = i * j
            print("%d * %d = %d"%(j,i,result),end = "\t")
            j = j + 1
        print("")
        i = i + 1
```
