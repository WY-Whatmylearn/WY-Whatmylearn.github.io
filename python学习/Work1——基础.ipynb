# 变量的格式化输出

```python
name1 = '小明' 
name2 = '小军'
print('我的名字叫 %s ，他的名字叫 %s,请多多关照！'%(name1,name2))
print('我的名字是%s'%'小明')

student_no = 1
print('我的学号是%06d'% student_no)

price = 9.00
weight = 5.00
money = price * weight
print('苹果单价为%0.2f元/斤，购买了%0.2f斤，需要支付%0.2f元'%(price,weight,money))

scale = 0.1
print('数据比例是%0.2f%%'%(scale * 100))
```
