# 元组的函数

```python
info = (1,2,3,1,2)
info.count(1)
```

```python
info.index(2)
```

# 元组的遍历

```python
for my_info in info:
    print(my_info)
```

# 元组的应用场景

```python
# 格式化字符串
info = ("小明",18,1.75)
print("他的名字是%s，今年%d岁,身高%.2f米"%info)
```

```python
info_str = "他的名字是%s，今年%d岁,身高%.2f米"%info
print(info_str)
```

# 元组和列表之间的转换

```python
a_list = [1,2,3]
a_tuple = tuple(a_list)
type(a_tuple)
```

```python
a2_list = list(a_tuple)
type(a2_list)
```
