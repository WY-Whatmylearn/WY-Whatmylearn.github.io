```python
class People():
    def __init__(self,name,weight):
        self.name = name
        self.weight = weight
        
    def __str__(self):
        return "%s%.2f"%(self.name,self.weight)

    def run(self):
        self.weight -= 0.5
        
    def eat(self):
        self.weight += 1
    
xiaoming = People("",75.0)
xiaoming.run()
xiaoming.eat()
print(xiaoming)
```
